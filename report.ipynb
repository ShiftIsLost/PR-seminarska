## 1. Vprašanje:  Razlike glede na izobraženost po občinah

Vprašanja katera občina je najbolj izobražena smo se lotili tako, da smo pridobil podatke o izobrazbi delovno aktivnega prebivalstva Slovenskih občin med letoma 2005 in 2021.
Na podlagi teh podatkov smo ustvaril grafe, ki so nam pokazali kakšni so deleži delovno aktivnega prebivalstva z visokošolsko, srednješolsko oziroma osnovnošolsko izobrazbo v vseh Slovenskih občinah.
Iz grafov je razvidno, da je občina z največjim deležem visokošolskega prebivalstva dolga leta bil Trzin, v zadnjih letih pa je to Kranjska Gora. Blizu je tudi Ljubljana. 
Med letoma 2005 in 2012 je bila občina z najmanjšim deležem prebivalstva z visokošolsko izobrazbo Hodoš. V letu 2021 je na predzadnjem mestu, Zavrč pa je zadnji.
Občina z največjim deležem prebivalstva s srednješolsko izobrazbo je bila med letoma 2005 in 2015 ter v letu 2020 in 2021 Osilnica, med letoma 2016 in 2019 pa je to Zavrč.
Po najaktualnejših podatkih je Ljubljana občina z najmanjšim deležem prebivalstva s srednješolsko izobrazbo. 
Občina z največjim deležem prebivalstva z osnovnošolsko izobrazbo je bil vsa leta (2005 - 2021) Hodoš, občina z najmanjšim deležem takšnega prebivalstva pa je trenutno Kranjska Gora.

Ker se nam grafi niso zdeli najbolj ustrezna vizualizacija podatkov, smo ustvaril tudi 'choropleth' zemljevid Slovenskih občin, ki je na bolj pregleden način prikazal podatke.
Tako imamo hiter in ustrezen pregled nad tem katere občine in predeli Slovenije imajo najmanjši ter največji delež določene stopnje izobrazbe. Prav tako je iz legende razvidno kako se spreminjajo deleži skozi leta - viden je strm padec deleža prebivalstva z osnovnošolsko izobrazbo in rast prebivalstva z visokošolsko izobrazbo. 
Iz podatkov sem ustvaril tudi grafe, ki prikazujejo trend deleža prebivalstva z določeno izobrazbo v petih izbranih Slovenskih mestih (Ljubljana, Maribor, Koper, Celje, Bled). 
Iz dobljenih grafov je razvidno, da delež prebivalstva z osnovnošolsko prebivalstvo v teh mestih pada, delež z visokošolsko izobrazbo pa raste. 
Delež prebivalstva z srednješolsko izobrazbo je v Kopru ter Celju ostal dokaj konstanten, v Bledu, Ljubljani in Maribor pa je upadel.

povezavo med deležem posamezne stopnje izobrazbe občine in deležem kmetov v občini. Iz pridobljenih podatkov je bilo razvidno da delež prebivalstva z osnovnošolsko izobrazbo občine signifikantno korelira z

z srednješolsko izobrazbo in deležem samozaposlenih kot kmetje ni bila signifikantna.

## 2. Vprašanje: Kakšna je razlika v akademskih dosežkih med spoloma?

Tukaj smo se osredotočili na primarno in sekundarno izobrazbo. Za primarno izobrazbo smo želeli pogledati, če je kateri izmed spolov v zaostanku in ali je to zelo opazno.

Pogledali smo razlike koliko fantov in deklic ponavljajo osnovni razred. Te razlike so prikazane kot procent vseh fantov/deklic razreda, katerega ponavljajo. Vidno je, da v povprečju ponavlja več fantov kot deklic s približnim razmerjem 2:3. Razlog, zakaj ponavlja več fantov, bi lahko pripisali temu, da se jih tudi v osnovno šolo vpiše več, vendar razlika pri vpisu je zelo majhna (nekaj procentov na več 1000 otrok).
V 9. razredu se pojavi nekaj zanimivega. Število ponavljavcev zelo pade nekje tudi za 10x. Tukaj se vidi, da so razlike med spoloma zelo majhne oziroma bolj podobne vpisnim razlikam za razred. To bi bilo lahko zaradi tega, ker so učitelji bolj tolerantni do zadnjega razreda.

Pri sekundarni stopnji izobraževanja smo najprej pogledali, koliko dijakov opravi to stopnjo. In opazili, da konča več žensk kot moških. Razvidno je tudi, da trend izobraženosti narašča pri obeh spolih.

Nato smo pogledali samo določena področja, da bi bolje razumeli, za kakšno izobrazbo se dijaki odločajo in kakšne so razlike v spolih na določenem področju.

Opazili smo, da se ženske odločajo bolj za socialne vede, kjer je več dela z ljudmi kot npr. zdravstvo in socialna varnost, umetnost in humanistika, Poslovne in upravne vede, pravo, transport, varnost, gostinstvo in turizem, osebne storitve in še področne, na katerem je največja razlika izobraževalne znanosti in izobraževanje učiteljev.

Moški so se odločali bolj za tehnične vede. To je jasno razvidno na dveh zelo tehničnih področjih tehnika, proizvodnje tehnologije in gradbeništvo in pri informacijskih in komunikacijskih tehnologijah.

Področje, ki je najbolj oravnovešeno je kmetijstvo, gozdarstvo, ribištvo in veterinarstvo, saj zajema širše področje poklicov, ki so privlačni obema spoloma.

## 3. Vprašanje: Razlike v temah glede na spol

Problematika vprašanja je kakšne teme so bolj zanimive kateremu spolu. Katere teme so raziskovali le moški, ali le ženske, katere teme niso močno obtežene le na en spol, torej jih lahko tretiramo kot bolj nevtralne.

Podatke smo dobili iz Open Science Slovenija: https://www.openscience.si/OpenData.aspx (različica 3 - 9.3.2021). Iz njih pridobimo imena avtorjev del. Težava pri podatkih je bila uporaba ločila oziroma delimiterja  '|' v csv datoteki, saj se '|' pojavi tudi v abstraktu in znotraj drugih besedil. 

Tej podatki ne vsebujejo spola avtorja, zato se jih določi glede na njihovo ime s kombinacijo podatkov iz SURS;

in uporabi se še knjižnica podatkov, ki so bili pridobljeni iz Facebook leaka, ki vsebuje imena in razmerje med spoloma, ki imajo to ime;

Nato skombiniramo oba dataseta, da dobimo skupno ujemanje imen in spola, tako nam izmed 8000 unikatnih imen ostane le še 513 nedoločenih imen:

V podatkovni set nastavimo spol, kjer je ime veljavno drugače ga izpustimo. Potrebno pa je tudi počistiti ID, torej kjer je ID niz, toda ne vsebuje drugih znakov kot cifre, se podatkovni tip spremeni v int, drugače se ga izbriše. Počiščena seta podatkov združimo.

Ustvarimo slovar, ki vsebuje slovar za vsak spol (moški, ženska, drugo). Slovar spola je oblike ključna beseda(keyword) => število pojavitev, s tem se sprehodimo po vsakem delu, in vsak podan keyword prištejemo spolu, ki je podan avtorju tega dela.

Iz tega slovarja izračunamo skupno število pojavitve keyworda, in še razmerje m/m+ž in ž/m+ž. Hitro je opazno, da so najpogostejše ključne besede splošne oznake za tip dela, torej "diplomska naloga", "magistrska dela", "UNI", ipd. zato se ustvari seznam teh besed in se jih izbriše iz preštetih pojavitev. Prav tako se odstrani ključne besede z manj kot 200 pojavitev, za večjo merodajnost, razen pri predstavitvah o 100% razmerju za en spol.

Iz podatkov smo izbrali 6 predstavitev oziroma grafov, kjer smo izpostavili 20 najbolj primernih ključnih besed. Grafi so sortirani naraščujoče, da so najbolj primerne ključne besede proti dnu.

Najprej smo se osredotočili na teme, ki so najbolj prisotne le pri enem spolu:

Opazimo, da je najbolj žensko utežena tema z vsaj 200 pojavitev * **opismenjevanje***, največkrat pojavljenja tema, z top 20 največjim razmerjem ženskih avtoric pa je * **zdravstvena vzgoja***.

Pri moških pa opazimo, da je najbolj utežena tema * **vgrajeni sistemi***, najpogostejša pa * **metoda končnih elementov***.

Poleg najbolj uteženih pa smo izbrali tudi najbolj uravnotežene oziroma nevtralne teme:

Teme smo sortirali po najboljižjem razmerju 50/50, izluščili 20 najbljižjih, in jih sortirali po številu pojavitev. Teme kjer je bilo razmerje točno 50% so le 4: * **Španija, elektronsko trženje, storilci, sredstva***. Zato smo razmerje omejili da je med 50,3% in 49,61%. V tem seznamu tem je najpogostejša tema * **ponudba***, zanimiva pojavitev pa je tudi tema * **razlike***.

Na splošno najbolj pogoste teme so seveda bolj splošne

S presenetljivo pogostostjo vidimo temo * **gradbeništvo***, daleč najbolj pogosta tema izven splošnih oznak naloge pa je * **Slovenija***.

Na koncu pa pregledamo še teme, ki so jih raziskovali le moški ali le ženske, pojavitev teh tem je manj, zato niso omejene na >200 pojavitev.

Najbolj pogosta tema, ki so jo raziskovale le ženske je * **prekmursko narečje***, zanimva pojavitev pa je tudi * **kletvice***.

Najbolj pogosta tema pri moških pa je * **električni stroji***. Opazimo pa, da je splošni trend, da je več pojavitev pri temah, ki so jih raziskovali le moški.
